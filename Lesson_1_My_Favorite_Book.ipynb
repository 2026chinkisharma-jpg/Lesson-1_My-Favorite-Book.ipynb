```python
book1_name = "To Kill a Mockingbird"
book1_author = "Harper Lee"
book1_pages = 281
book1_year = 1960
book1_copies = 10
book1_languages = "English, Spanish"
book1_price = 9.99
```

```python
print("\nBook:", book1_name)
print("Author:", book1_author)
print("Pages:", book1_pages)
print("Year of Publishing:", book1_year)
print("Copies:", book1_copies)
print("Available in:", book1_languages)
print("Price: $", round(book1_price, 2))
```
